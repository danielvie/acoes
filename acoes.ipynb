```python
import pandas as pd
import numpy as np
import re
```

```python
# importando portifolio
P = pd.read_csv('Portfolio-03-31-2022.csv')
P = P[['Id', 'Symbol', 'Name', 'Portfolio', 'Currency', 'Quantity', 'Cost Per Share', 'Date', 'TimeOfDay', 'Type', 'Notes        ']]
P.columns = ['id', 'symbol', 'name', 'portfolio', 'currency', 'qnt', 'cost', 'date', 'time', 'type', 'notes']
```

```python
# encontrando coluna 'corretora'
tam = len(P);

corretora = ['',]*tam
for idx, row in P.iterrows():
    
    # ignorando valores `float`
    if type(row.notes) == float:
        continue
    
    # lendo nota em lower case
    nome = row.notes.lower()
    
    # inter
    if 'inter' in nome:
        corretora[idx] = 'inter'
        continue
    
    # modal
    if 'modal' in nome:
        corretora[idx] = 'modal'
        continue
        
    # clear
    if 'clear' in nome:
        corretora[idx] = 'clear'
        continue
    
    # c6
    if 'c6' in nome:
        corretora[idx] = 'c6'
        continue
    
    # atribuindo restante para 'inter'
    corretora[idx] = 'inter'
    
P['corretora'] = corretora
```

```python
# filtrando Bovespa
Bovespa = P[(P.portfolio == 'Bovespa') & (~P.qnt.isna())]
Bovespa
```

```python
def get_symbol(P, acao):
    S = P.symbol.unique()
    
    S_out = []
    for s in S:
        if acao.upper() in s:
            S_out.append(s)
    
    return S_out

get_symbol(P, 'trad3')
```

```python
def get_transacoes_from_port(P, acao):
    
    # lendo simbolo acao
    symbol = get_symbol(P, acao)[0]
    
    # filtrando acao
    Pi = P[P.symbol == symbol]

    return Pi

T = get_transacoes_from_port(Bovespa, 'TRAD3')
```
